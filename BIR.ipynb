# Rate monotonic scheduler

This program was created for the embedded systems class in the Budapest University of Technology and Economics. (BME - Beágyazott információs rendszerek házi feladat)
It is emulating a scheduler for some example Tasks. 

Every task has a name, a periodic time, a calculation time, and a starting phaze. The scheduler alwyas starts at 0. 

```python
class Task:
    def __init__(self, period, calculationTime, startingPhaze, name):
        self.period = period
        self.calculationTime = calculationTime
        self.startingPhaze = startingPhaze
        self.startingTime = startingPhaze
        self.timeRemain = calculationTime
        self.name = name

    name = 0
    period = 0
    calculationTime = 0
    startingPhaze = 0

    timeRemain = 0
    
    def refresh(self, time):
        if (time - self.startingPhaze) % self.period == 0:
            self.timeRemain = self.calculationTime
            self.startingTime = time
            return True
    
    def run(self):
        self.timeRemain = round(self.timeRemain - 0.1, 2)
        
    def print(self):
        print("Periódus,", self.period, ",Számítási idő,", self.calculationTime, ",Fázis,", self.startingTime)


p1 = 2
p2 = 5
p3 = 3

scheduler = Task(2*p1, 0.5, 0, "S")

t1 = Task(8*p1, 2*p1, p3*p1+0.1, "T1")
t2 = Task(16*p1, 3*p1, p2*p1+0.2, "T2")
t3 = Task(32*p1, 4*p1, p1*p1+0.3, "T3")
t4 = Task(64*p1, 15*p1, 0.4, "T4")


tasks = [t1, t2, t3, t4]

def sortByPeriodicity(task):
    return task.period

tasks.sort(key=sortByPeriodicity)

for i in range(0, len(tasks)):
    tasks[i].print()
scheduler.print()
```

```python
output = ""

time = 0
startRunTime = 0
taskToRun = None
taskWillRun = None
schedulerStartTime = 0
while(time <= 128+4):
    # refresh the tasks if new period starts
    for task in tasks:
        task.refresh(time)


    #1 schudler start running
    if((time - scheduler.startingPhaze) % scheduler.period == 0):
        schedulerStartTime = time
        desided = False
        for i in range(0, len(tasks)):
            task = tasks[i]
            if (not desided) and time >= task.startingPhaze and task.timeRemain > 0:
                if(time == 8):
                    print()
                if task == taskToRun and task.timeRemain <= scheduler.calculationTime:
                    continue
                
                taskWillRun = task
                desided = True
            

    #2 scheduler updates the running task
    elif(time == schedulerStartTime + scheduler.calculationTime):
        scheduling = False
        taskToRun = taskWillRun
        schedulerStartTime = 0
        taskWillRun = None


    #3 the running task is running
    if(taskToRun != None and taskToRun.timeRemain > 0):
        taskToRun.run()
        output += taskToRun.name
    else:
        output += "X"
        
    output += ","
    
    

    time += 0.1
    time = round(time, 2)

# save the output
output = output[:-1]
file = open("output.txt", "w")
file.write(output)
file.close()
```

```python
# Save the scheduler timing

time = 0
schedlerOut = ""
scheduling = False
schedulerStartTime = 0
while(time <= 128+4):
    if((time - scheduler.startingPhaze) % scheduler.period == 0):
        scheduling = True
        schedulerStartTime = time
    if scheduling:
        schedlerOut += "S"
        if(time == schedulerStartTime + scheduler.calculationTime):
            scheduling = False
            schedlerOut = schedlerOut[:-1]
    time += 0.1
    time = round(time, 2)
    schedlerOut += ","

file = open("scheduler.txt", "w")
file.write(schedlerOut)
file.close()
```
